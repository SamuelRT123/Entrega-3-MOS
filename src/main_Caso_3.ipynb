#  Calibración y experimentación Caso 3 Proyecto A

```python
import random

from data_loader import load_instance
from ga_algorithm import GeneticAlgorithm
from verification import export_verification
import time
import pandas as pd

```

```python
# SemillaS PARA LAS TRES PRUEBAS
SEED1 = 42
SEED2 = 110
SEED3 = 172
```

```python

# PARAMETROS GLOBALES DEL GA 

# Tamaño de población
POP_SIZE = 50

# Número de generaciones
N_GENERATIONS = 200

# Probabilidades de operadores
P_CROSS = 0.8      # probabilidad de cruce
P_MUT = 0.2        # probabilidad de mutación
```

### PRUEBA 1

```python
resultados1 = []
```

```python
# Fijar semilla global (por si hay otros randoms fuera del GA)
random.seed(SEED1)

instance_path = "../data/Proyecto_Caso_3"
config = load_instance(instance_path)


# CONSTRUCCIÓN DEL GA
ga1 = GeneticAlgorithm(
    config=config,
    pop_size=POP_SIZE,
    generations=N_GENERATIONS,
    crossover_rate=P_CROSS,
    mutation_rate=P_MUT,
    seed=SEED1,
)

# EJECUCIÓN
t0 = time.process_time()
best1, history = ga1.evolve()
cpu1 = time.process_time() - t0

print("\nMejor solución encontrada:")
print("Costo:", best1.cost)
print("Factible:", best1.is_feasible)
print("\nParámetros GA usados:")
print(f"  Semilla           : {SEED1}")
print(f"  Población         : {POP_SIZE}")
print(f"  Generaciones      : {N_GENERATIONS}")
print(f"  P(cruce)          : {P_CROSS}")
print(f"  P(mutación)       : {P_MUT}")
print(f"  Tiempo CPU        : {cpu1:.4f} s")

resultados1.append({
    "semilla": SEED1,
    "objetivo": best1.cost,
    "factible": best1.is_feasible,
    "cpu_time": cpu1,
})
```

```python
import folium
import math

def make_solution_map(config, df_verif):

    depot_code_cfg = str(config.depot.code).strip().upper()
    depot_code_csv = str(df_verif["DepotId"].iloc[0]).strip().upper()


    clients_norm = {}
    for c in config.clients.values():
        key = str(c.code).strip().upper()
        clients_norm[key] = c

    all_lat = [config.depot.lat] + [c.lat for c in config.clients.values()]
    all_lon = [config.depot.lon] + [c.lon for c in config.clients.values()]

    center_lat = sum(all_lat) / len(all_lat)
    center_lon = sum(all_lon) / len(all_lon)

    m = folium.Map(
        location=[center_lat, center_lon],
        zoom_start=12,
        tiles="OpenStreetMap"
    )

    folium.Marker(
        location=[config.depot.lat, config.depot.lon],
        popup=f"Depot {config.depot.code}",
        icon=folium.Icon(color="red", icon="home", prefix="fa")
    ).add_to(m)


    for client in config.clients.values():
        folium.CircleMarker(
            location=[client.lat, client.lon],
            radius=4,
            popup=f"{client.code} - Demanda: {client.demand}",
            color="blue",
            fill=True,
            fill_opacity=0.7,
        ).add_to(m)

    colors = [
        "blue", "green", "purple", "orange", "darkred", "cadetblue",
        "red", "black", "darkblue", "darkpurple"
    ]


    for idx, row in df_verif.iterrows():
        vehicle_id = row["VehicleId"]
        seq = row["RouteSequence"]

        if isinstance(seq, float) and math.isnan(seq):
            continue

        nodes = [s.strip() for s in str(seq).split("-") if s.strip() != ""]

        coords = []
        for code in nodes:
            code_norm = str(code).strip().upper()

            if code_norm in (depot_code_csv, depot_code_cfg):
                n = config.depot
            elif code_norm in clients_norm:
                n = clients_norm[code_norm]


            coords.append((n.lat, n.lon))

        if not coords:
            continue

        col = colors[idx % len(colors)]

        folium.PolyLine(
            coords,
            color=col,
            weight=4,
            opacity=0.8,
            tooltip=f"Vehículo {vehicle_id}"
        ).add_to(m)

    return m
```

```python
solution_map = make_solution_map(config, df)
solution_map
```
